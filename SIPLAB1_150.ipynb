In this experiment we will learn to create and plot signals using numpy, matplot.plyplot and other python libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
a = 1
Fs = 5000
f = 10
d=15
```

```python
#UNIT STEP
t = np.arange(0, 1, 1/Fs)

# Unit step function
u = np.where(t >= 0, a, 0)

# Plotting
plt.plot(t, u)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Unit Step Function')
plt.grid(True)
plt.show()
```

Inference:

From the above code we learnt how to create a unit step signal and observed its structure.

```python
#UNIT RAMP
t = np.arange(0, 1, 1/Fs)
ramp = np.where(t >= 0, a*t, 0)
plt.plot(t, ramp)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Unit Ramp Signal')
plt.show()
```

Inference:

From the above code we learnt how to create a unit ramp signal and observed its structure.

```python
#SINE
duration = 10
f = 15/duration

t = np.linspace(0, duration, Fs*duration)

sig_fun = np.sin(2*np.pi*t*f)

plt.plot(t,sig_fun)
plt.title("Sine wave" )
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

Inference:

From the above code we learnt how to create a unit sine wave signal and observed its structure with respect to chang in frequency, duration and sampling rate.

```python
#COSINE
t = np.linspace(0, duration, Fs*duration)

sig_fun = np.cos(2*np.pi*t*f)

plt.plot(t,sig_fun)
plt.title("Cosine")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

Inference:

From the above code we learnt how to create a unit cos wave signal and observed its structure with respect to chang in frequency, duration and sampling rate.

```python
#SAWTOOTH
from scipy import signal
t = np.linspace(0,14,int(Fs * d))

sawtooth_wave = signal.sawtooth(2 * np.pi * f * t)

plt.plot(t, sawtooth_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sawtooth Signal')
plt.show()
```

Inference:

```python
#SQUARE
from scipy import signal
t = np.linspace(0,14,int(Fs * d))
square_wave = signal.square(2 * np.pi * f * t)

plt.plot(t, square_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square Signal')
plt.show()
```

Inference:

```python
#TRIANGLE
t = np.linspace(0,14,int(Fs * d))
triangle_wave = signal.sawtooth(2 * np.pi * f * t, width=0.5)
plt.plot(t, triangle_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Triangle Signal')
plt.show()
```

Inference:

```python
#EXPONENTIAL
t = np.linspace(0, duration,Fs*duration)

sig_fun = (2)**t
plt.plot(t,sig_fun)
plt.title("Exponential function")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

Inference:

From the above code we learnt how to plot an exponential equation in matplot and see it visualy how exponentialy the difference in the values affets the output.

```python
#Exponential function (decreasing)
t = np.linspace(0, duration,Fs*duration)

sig_fun = (1/2)**t

plt.plot(t,sig_fun)
plt.title("Exponential function")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

Inference:

From the above code we learnt how to plot an exponential equation in matplot and see it visualy how exponentialy the difference in the values affets the output.

```python
#GAUSSIAN
x = np.linspace(-10, 10, 1000)
t = np.linspace(-duration, duration, int(Fs * duration))
mu=0
sigma=1
sig_fun = a * np.exp(-((t - mu)**2) / (2 * sigma**2))

plt.plot(t, sig_fun)
plt.title("Gaussian Function")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

Inference:

From the above code we learnt how to plot gaussian function using matplot and see the characteristic symmetric "bell curve".

```python
#SINC
t = np.linspace(-duration, duration, int(Fs * duration))

sig_fun = np.sinc(t * f)  # or sig_fun = np.sin(np.pi * t * f) / (np.pi * t * f)

plt.plot(t, sig_fun)
plt.title("Sinc Function")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

Inference:

From the above code we learnt how to plot sinec function using matplot and By plotting the sinc function, we can visually observe its characteristic lobes and gain insights into its behavior.

```python
#SUM
from scipy import signal
t = np.linspace(-duration, duration, int(Fs * duration))

sawtooth = a*signal.sawtooth(2 * np.pi * f * t,width = 0.4)

sin =  np.sin(2*np.pi*t*f)

mix= sin+sawtooth

plt.plot(t, sin, label='Sine')
plt.plot(t, sawtooth, label='Saw tooth')
plt.plot(t, mix, label='Sine + sawtooth')

plt.title("Sin + sawtooth")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.legend()
plt.show()
```

Inference:

From the above code we learnt how to add two signal and observed how the structure of waveform changes .

```python
#SUBTRACT
from scipy import signal
t = np.linspace(-duration, duration, int(Fs * duration))

sawtooth = a*signal.sawtooth(2 * np.pi * f * t,width = 0.4)

sin =  np.sin(2*np.pi*t*f)

mix= sin-sawtooth

plt.plot(t, sin, label='Sine')
plt.plot(t, sawtooth, label='Saw tooth')
plt.plot(t, mix, label='Sine - sawtooth')

plt.title("Sin - sawtooth")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.legend()
plt.show()
```

Inference:

From the above code we learnt how to subtract two signal and observed how the structure of waveform changes .

```python
#MULTIPLICATION
from scipy import signal
t = np.linspace(-duration, duration, int(Fs * duration))

sinc = np.sinc(t * f)

sin =  np.sin(2*np.pi*t*f)

mix= sin*sinc

plt.plot(t, sin, label='Sine')
plt.plot(t, sinc, label='Sinc')
plt.plot(t, mix, label='Sine * Sinc')

plt.title("Sin * Sinc")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.legend()
plt.show()
```

Inference:

From the above code we learnt how to multiply two signal and observed how the structure of waveform changes .

```python
#DIVISION
from scipy import signal
t = np.linspace(-duration, duration, int(Fs * duration))

sinc = np.sinc(t * f)

sin =  np.sin(2*np.pi*t*f)

mix= sin/sinc

plt.plot(t, sin, label='Sine')
plt.plot(t, sinc, label='Sinc')
plt.plot(t, mix, label='Sine / Sinc')

plt.title("Sin * Sinc")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.legend()
plt.show()
```
